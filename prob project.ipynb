```python
 
d2['date'] = pd.to_datetime(d2.index,format='%Y%m%d')
d2['year'] = pd.DatetimeIndex(d2.index).year
d2['month'] = pd.DatetimeIndex(d2.index).month
d3['date'] = pd.to_datetime(d3.index,format='%Y%m%d')
d3['year'] = pd.DatetimeIndex(d3.index).year
d3['month'] = pd.DatetimeIndex(d3.index).month
d1['date'] = pd.to_datetime(d1.index,format='%Y%m%d')
d1['year'] = pd.DatetimeIndex(d1.index).year
d1['month'] = pd.DatetimeIndex(d1.index).month
```

```python
d3['btc_return'] = d3['close'].pct_change()
d1['sp_return'] = d1['close'].pct_change()
d2['gold_return'] = d2['gold_usd'].pct_change()
```

```python
d3.head()
```

```python

bitcoin_month = d3.groupby(['year','month'])['btc_return'].sum()
bitcoin_month= pd.DataFrame(bitcoin_month)
sp500_month = d1.groupby(['year','month'])['sp_return'].sum()
sp500_month= pd.DataFrame(sp500_month)
gold_month = pd.DataFrame(d2.groupby(['year','month'])['gold_return'].sum())
```

```python
monthly_prices = bitcoin_month.merge(sp500_month, left_index = True,right_index=True, suffixes = ['_btc', '_sp500'])
monthly_prices = monthly_prices.merge(gold_month, left_index = True,right_index=True)
monthly_prices
```

```python
std_figure_size = (14, 7)
plt.figure(figsize=std_figure_size)
plt.plot(monthly_prices["btc_return"].values)
plt.plot(monthly_prices["sp_return"].values)
plt.plot(monthly_prices["gold_return"].values)
plt.legend(["bitcoin","s&p","gold"])
plt.xlabel("Months")
plt.ylabel("Return %")
plt.title("Monthly Returns")
plt.show()
```

```python
inf = d2.groupby(['year'])['cpi_us'].mean()
inf = pd.DataFrame(inf)
inf['inflation'] = inf['cpi_us'].pct_change() 
inf['inflation']=inf['inflation'].fillna(0)
```
